```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/data/kaggle/tabular-playground-series-jan-2022/train.csv")
```

```python
#
```

```python
df_date = df.groupby(["date"]).agg({"num_sold": "sum"})
```

```python
df_date2 = df_date.reset_index()
df_date2["date"] = pd.to_datetime(df_date2["date"], format="%Y-%m-%d")
```

```python
df_date2.head()
```

```python
sns.lineplot(x="date", y="num_sold", data=df_date2)
```

```python
df_country = df.groupby(["date", "country"]).agg({"num_sold": "sum"})
```

```python
df_country.head()
```

```python
sns.lineplot(x="date", y="num_sold",hue="country", data=df_country)
```
